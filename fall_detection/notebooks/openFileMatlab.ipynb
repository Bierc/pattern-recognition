```python
import scipy.io as sio
import matplotlib.pyplot as plt
```

```python
mat = sio.loadmat("../data/sameWindowData/1.1.1.01.mat")
data = mat["newData"]
```

```python
# The 11 columns are: time, accelerometer (X, Y e Z), gyroscope (X, Y e Z), magnetometer (X, Y e Z), and a zero column
print(data.shape)
print(data)
```

```python
x = data[:, 0] / 100 # divide by 100 to convert in seconds, since Fs = 100Hs

plt.plot(x, data[:, 2]) # Accelerometer Y axis
plt.xlabel("Time [s]")
plt.ylabel("Value")
plt.title("Accelerometer Y axis")
plt.tight_layout()
plt.show()
```
